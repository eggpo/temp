```r
flight.1 <- hflights  %>% group_by(Origin, Dest)  %>%  
summarise(avg_Delay=round(mean(DepDelay, na.rm=T), 3), 
          avg_ArrDelay=round(mean(ArrDelay, na.rm=T),3))
```

```r
str(flight.1)
```

```r
class(flight.1)
```

```r
# https://stackoverflow.com/questions/16941111/r-cannot-melt-data-frame
# dplyr issue
flight.1 <- as.data.frame(flight.1)
```

```r
#https://github.com/hadley/reshape/issues/63
# variable name issue
# https://stackoverflow.com/questions/6979917/how-to-unload-a-package-without-restarting-r
detach("package:reshape", unload=TRUE)
```

```r
library(reshape2)
```

```r
flight.melt <- melt(flight.1, id="Origin", measure=c("avg_Delay", "avg_ArrDelay"), 
                    variable.name="Delay_type", value.name="Delay_time",
                    na.rm=T)
```

```r
class(flight.melt)
```

```r
tail(flight.melt)
```

```r
ggplot(flight.melt, aes(x=Delay_time, fill=Delay_type, alpha=0.0000001)) + geom_density(color="lightgrey", size=0.0000001) 
+ facet_wrap( ~ Origin, nrow=2)
```
