# dplyr Basics

* 조건에 맞는 row값을 찾습니다.(filter())

```r
grade <- c("A+", "A", "B+", "B", "C+", "C", "D+", "D", "F")
```

```r
set.seed(1234)
class_grade <- sample(grade, 100, replace = T)
```

```r
class_grade
```

```r
match("A", grade)
```

```r
switch(match("A+", grade), 4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0)
```

```r
calc_grade <- function(x){
    return(switch(match(x, grade), 4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0))
}
```

```r
calc_grade("B+")
```

```r
for(x in class_grade){
    print (calc_grade (x))
}
```

```r
mean(unlist(lapply(class_grade, calc_grade)))
```
