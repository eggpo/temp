# 패키지 

## 패키지가 설치되는 경로 알아보기
* 패키지가 설치되어 있는 폴더의 경로는 .libPaths()를 통해서 확인할 수 있다.

```r
.libPaths()
```

## 설치되어 있는 패키지 확인하기 
* 설치되어 있는 패키지가 어떤 것이 있는지 확인하고 싶다면 아래 명령어를 실행하여 확인한다.

```r
installed.packages()
```

## 패키지의 자세한 내용을 확인하고 싶다면 library(help=패키지명)을 실행한다.

```r
library(help='ggplot2')
```

## 패키지 설치하기 
* 패키지 설치하는 방법은 instll.packages()를 호출하여 설치하고 싶은 패키지명을 괄호안에 적는다. 
* 윈도우 Jupyter에서는 repos를 설정행줘야 정상적으로 실행된다. 
* 한번에 여러개의 패키지를 설치하고 싶다면 아래와 같이 c() 함수를 사용해서 여러 패키지명을 적어주면 된다.

    > install.packages(c("ggplot2", "dplyr"))

```r
install.packages("dplyr",  repos='http://cran.us.r-project.org')
```

### 패키지가 설치되어 있지 않다면 자동으로 패키지 설치하기

```r
if(!requireNamespace("plyr")) install.packages('plyr', repos='http://cran.us.r-project.org')
```

## 패키지 업데이트 하기 
* update.packages()를 실행하면 업데이트가 가능한 패키지들은 모두 대상이 되고, 개별로 업데이터를 하고 싶다면 update.packages(패키지이름)을 실행

```r
update.packages(repos='http://cran.us.r-project.org')
```

## 패키지 제거
* 제거하고 싶은 패키지는 remove.packages()를 사용하면 된다. 여러 패키지를 삭제하고 싶다면 c()를 이용한다.

```r
remove.packages(c("plyr", 'dplyr'))
```

## 패키지 로드하기
* 설치된 패키지를 사용할때는 library()를 사용해서 로드시키면 된다.

```r
library(dplyr)
```

## 패키지 언로드하기
* 같은 namespace에 함수명이 겹치서 특정 패키지를 언로드(떼기)를 할 필요가 있을 때 detach()를 사용한다.

```r
detach("package:dplyr", unload=T)
```

## 현재 R 세션에서 어떤 패키지들이 사용되고 있는지 확인하기
* search() 함수는 어떤 패키지들이 사용되고 있는지 출력

```r
search()
```
