```r
apt.seoul.dong  %>% 
group_by(구)  %>% 
summarise(평균거래금액 = mean(거래금액))   %>% 
arrange(desc(평균거래금액))
```

```r
apt.seoul.dong  %>% 
group_by(구)  %>% 
summarise(
         apt.max= max(거래금액),
         apt.min= min(거래금액),
         apt.mean = mean(거래금액),
         apt.count = n())  %>% 
arrange(desc(apt.mean))
```

```r
apt.seoul.dong  %>% 
group_by(구, 동)  %>% 
summarise(
         apt.max= max(거래금액),
         apt.min= min(거래금액),
         apt.mean = mean(거래금액),
         apt.count = n())  %>% 
arrange(desc(apt.count))
```

```r
apt.seoul.dong  %>% 
group_by(구, 동, 단지명)  %>% 
summarise(
         apt.max= max(거래금액),
         apt.min= min(거래금액),
         apt.mean = mean(거래금액),
         apt.count = n())  %>% 
arrange(desc(apt.max))  %>% 
head(100)
```

```r
apt.seoul.dong  %>% 
#filter( 구=="강남구" | 구=="서초구" | 구=="송파구")  %>% 
filter(구 %in% c("강남구", "서초구", "송파구"))  %>% 
group_by(구,동)  %>% 
summarise(
         apt.max= max(거래금액),
         apt.min= min(거래금액),
         apt.mean = mean(거래금액), 
         apt.median = median(거래금액),
         apt.count = n())  %>% 
arrange(desc(apt.mean))
```

```r
kangnam <- apt.seoul.dong  %>% 
filter(구 %in% c("강남구", "서초구", "송파구"))  %>% 
group_by(계약년월)  %>% 
summarise(
         apt.max= max(거래금액),
         apt.min= min(거래금액),
         apt.mean = mean(거래금액), 
         apt.median = median(거래금액))  %>% 
arrange(desc(apt.mean))
```
